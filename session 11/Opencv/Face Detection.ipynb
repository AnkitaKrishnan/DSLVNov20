```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = plt.imread("./group_photo.jpg")
```

```python
img.shape
```

```python
plt.imshow(img)
```

```python
model = cv2.CascadeClassifier('./haarcascade_frontalface_alt.xml')
```

```python
faces = model.detectMultiScale(img, 1.2)
```

```python
faces
```

```python
faces.shape
```

```python
for face in faces:
    x,y,w,h = face
    img = cv2.rectangle(img, (x, y), (x+w, y+h), (0,255,0), 4 )
```

```python
plt.imshow(img)
```
