```python
import numpy as np
import matplotlib.pyplot as plt
```

## Line Plot

```python
x = np.arange(10)
y1 = x**2

y2 = 2*x + 3
```

```python
y1
```

```python
y2
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x, y1, linestyle="dashed",  color="green", marker="o", markersize= 8, label="iPhone", )
plt.plot(x, y2, color="red", marker="x", label="Android")
plt.xlabel("Time")
plt.ylabel("Prices")
plt.title("Prices Vs. Time")
plt.legend() # -> prints label
# plt.xlim(-2, 10)
# plt.ylim(-10, 30)
plt.xticks(np.arange(0,10, 0.5))
plt.show()
```

## Scatter Plot

```python
plt.scatter(x, y1)
plt.scatter(x, y2)
plt.show()
```

### Bar Graph

```python
plt.bar?
```

```python
x = np.array([2017,2018,2019,2020])
car1 = np.array([40, 18, 24, 36])
car2 = np.array([30, 32, 14, 40])
```

```python
plt.bar(x, car1, width=0.3)
plt.bar(x+0.3, car2, width=0.3)
plt.xticks(range(2017, 2021,1))
plt.show()
```
